## Joint probability distributions

A joint probability distribution is simmilar to a probability distribution, however, there are two variables involved, instead of one.

A table is created that shows the probability for both random variables to reach a certain value. For example

$X = 1$ and $Y = 2$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generate random data for two variables
np.random.seed(42)
x = np.random.randint(0, 5, 100)
y = np.random.randint(0, 5, 100)

# Joint frequency table (normalized for probability)
joint_table = pd.crosstab(x, y, normalize=True)

# Print joint probability table
print("Joint Probability Distribution Table:")
print(joint_table)

# Visualize joint probability distribution
plt.figure(figsize=(8, 6))
sns.heatmap(joint_table, annot=True, cmap="Blues", cbar=True)
plt.title("Joint Probability Distribution")
plt.xlabel("Variable Y")
plt.ylabel("Variable X")
plt.show()
```

## Correlation and Dependence

A correlation quantifies the relationship between two variables.
- -1 indicates a negative correlation
- 0 indicates no correlation at all
- 1 indicates a positive correlation

The following code will provide an example of this

```python
# Generate two correlated variables
mean = [0, 0]
cov = [[1, 0.8], [0.8, 1]]  # High positive correlation
data = np.random.multivariate_normal(mean, cov, 500)
x_corr, y_corr = data[:, 0], data[:, 1]

# Calculate Pearson correlation coefficient
correlation = np.corrcoef(x_corr, y_corr)[0, 1]
print(f"Pearson Correlation Coefficient: {correlation:.2f}")

# Scatterplot of correlated variables
plt.figure(figsize=(8, 6))
plt.scatter(x_corr, y_corr, alpha=0.6, color="purple")
plt.title(f"Scatter Plot of Correlated Variables\nCorrelation = {correlation:.2f}")
plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.grid()
plt.show()
```

## Random Samples

A random sample is a subset of random data points chosen from a larger population or probability distribution.

The sample is collected at random to ensure that the subset still represents the underlying population as best as it can.

The following code demonstrates an example of random samples.

```python
# Generate random samples from a normal distribution
sample_size = 1000
random_samples = np.random.normal(loc=0, scale=1, size=sample_size)

# Plot histogram of random samples
plt.figure(figsize=(8, 6))
plt.hist(random_samples, bins=30, alpha=0.7, color="green", density=True)
plt.title("Histogram of Random Samples from Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid()
plt.show()

# Statistical summary of the samples
mean_sample = np.mean(random_samples)
std_sample = np.std(random_samples)
print(f"Sample Mean: {mean_sample:.2f}")
print(f"Sample Standard Deviation: {std_sample:.2f}")
```
